```
import numpy as np
```

```
#Coordinate ordering
A=1
B=6
C=4
D=5
E=2
F=3

Letters = ["A","B","C","D","E","F"]
Order = [A,B,C,D,E,F]

Assignments = {"A":A,"B":B,"C":C,"D":D,"E":E,"F":F}

#Distances
AE=9.5   # d_1,2
AF=10.65 # d_1,3
AC=9.84  # d_1,4
AD=5.13  # d_1,5
AB=3.9   # d_1,6

EF=10.92 # d_2,3
EC=14    # d_2,4
ED=5.3   # d_2,5
EB=11.1  # d_2,6

FC=4.98  # d_3,4
FD=11.8  # d_3,5
FB=8.05  # d_3,6

CD=13    # d_4,5
CB=6.12  # d_4,6

DB=8.15  # d_5,6

Distances = {"AE":AE, "AF":AF, "AC":AC, "AD":AD, "AB":AB, "EF":EF, "EC":EC, \
             "ED":ED, "EB":EB, "FC":FC, "FD":FD, "FB":FB, "CD":CD, "CB":CB, \
             "DB":DB}
```

```
#Distance matrix
len(Assignments)
D=np.zeros((len(Assignments),len(Assignments)))

#populate the matrix
for i in range(D.shape[0]):
  for j in range(D.shape[1]):
    if j>i:
      Letter1=Letters[Order.index(i+1)]
      Letter2=Letters[Order.index(j+1)]
      Distance = Letter1+Letter2
      D[i,j]=Distances[Distance]
    elif j<i:
      Letter1=Letters[Order.index(j+1)]
      Letter2=Letters[Order.index(i+1)]
      Distance = Letter1+Letter2
      D[i,j]=Distances[Distance]

#print(D)
```
